```python
import random
import pandas as pd
from faker import Faker
from tqdm import tqdm
fake = Faker()

TIPO_PERSONA = ['ESTUDIANTE', 'EMPLEADO', 'EGRESADO', 'PROFESOR', 'INVITADO', 'PADRE']
TIPO_PERIODO = ['DIA', 'SEMANA', 'MES', 'SEMESTRE']
TIPO_OPERADOR = ['HOTEL', 'HOSTAL', 'PERSONA_NATURAL', 'VIVIENDA_UNIVERSITARIA']
TIPO_ALOJAMIENTO = ['SUITE', 'SEMISUITE', 'HOSTAL_COMPARTIDO', 'HABITACION_ESTANDAR', 
                    'HABITACION_INDIVIDUAL', 'HABITACION_COMPARTIDA', 'VIVIENDA_COMPARTIDA', 
                    'VIVIENDA_COMPLETA', 'APARTAMENTO_NO_AMOBLADO', 'APARTAMENTO_AMOBLADO']
TIPO_SERVICIO = ['INTERNET', 'WIFI', 'TV_CABLE', 'RECEPCION', 'ASEO', 'MENAJE', 'RESTAURANTE', 'GIMNASIO', 'MESA',
                 'COCINA', 'BAÑERA', 'PISCINA', 'PARQUEADERO', 'ELECTRICIDAD', 'TELEFONO', 'AGUA_CALIENTE', 'SALA',
                 'BAÑO_PRIVADO', 'BAÑO_COMPARTIDO', 'ADMINISTRACION', 'APOYO', 'SALA_SOCIAL', 'SALA_ESTUDIO']
```

```python
df = []

for i in tqdm(range(100_000)):
    df.append({
        "cedula": i,
        "nombre": fake.name(),
        "correo": fake.email().replace("@", "@" + str(i)),
        "contrasena": "password",
        "tipo": random.choice(TIPO_PERSONA),
        "alohandes": "Universidad de los Andes",
    })

pd.DataFrame(df).to_csv("Cliente.csv", index=False)
del df
```

```python
df = []

for i in tqdm(range(100_000)):
    df.append({
        "id_convenio": i,
        "nombre": fake.company()[:25],
        "correo": fake.email().replace("@", "@" + str(i)),
        "contrasena": "password",
        "tipo_operador": random.choice(TIPO_OPERADOR),
        "alohandes": "Universidad de los Andes",
    })

pd.DataFrame(df).to_csv("Operador.csv", index=False)
del df
```

```python
df = []

for i in tqdm(range(200_000)):
    df.append({
        "id_oferta": i,
        "tipo": random.choice(TIPO_ALOJAMIENTO),
        "costo": fake.pyint(min_value=1),
        "capacidad": fake.pyint(min_value=1),
        "periodicidad": random.choice(TIPO_PERIODO),
        "tiempo_minimo": fake.pyint(min_value=0),
        "ubicacion": fake.city(),
        "tamano": fake.pyint(min_value=1),
        "retiro": None,
        "operador": fake.pyint(min_value=0, max_value=100_000-1),
    })

pd.DataFrame(df).to_csv("OfertaAlojamiento.csv", index=False)
del df
```

```python
df = []

for i in tqdm(range(400_000)):
    df.append({
        "id_oferta": i,
        "creacion": fake.date(),
        "inicio": fake.date(),
        "periodos": fake.pyint(min_value=1),
        "costo": fake.pyint(min_value=1),
        "cliente": fake.pyint(min_value=0, max_value=100_000-1),
        "oferta": fake.pyint(min_value=0, max_value=200_000-1),
    })

pd.DataFrame(df).to_csv("Reserva.csv", index=False)
del df
```

```python
df = []

for i in tqdm(range(200_000)):
    df.append({
        "id_servicio": i,
        "nombre": fake.text(max_nb_chars=25),
        "descripcion": fake.text(max_nb_chars=25),
        "costo": fake.pyint(min_value=0),
        "tipo": random.choice(TIPO_SERVICIO),
        "oferta": fake.pyint(min_value=0, max_value=200_000-1),
    })

pd.DataFrame(df).to_csv("Servicio.csv", index=False)
del df
```

```python
df = []
already = set()

i = 0
pbar = tqdm(total=200_000)
while i < 200_000:
    servicio = fake.pyint(min_value=0, max_value=200_000-1)
    reserva = fake.pyint(min_value=0, max_value=400_000-1)
    tupla = (servicio, reserva)
    if tupla in already:
        continue
    df.append({
        "servicio": servicio,
        "reserva": reserva,
    })
    i += 1
    already.add(tupla)
    pbar.update(1)

pd.DataFrame(df).to_csv("ServicioAdquirido.csv", index=False)
del df
del already
```
